# Scatter Plots
The scatter plot is a mainstay of statistical visualization. It depicts the join distribution two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relation between them.

# There are two ways to draw a scatterplot in seaborn:
1. relplot() [Figure level function] / relation plot
2. scatterplot() [Axex level function]

As we will see, these visualization can be quite helpful because they use simple and easily understandable representation of data even on complex data by mapping up to three additional variables using the semantics of hue, size and style.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

# Plot a scatter plot between total_bill and tip

```python
# plot here
sns.relplot(x='total_bill', y='tip', data=data, kind='scatter')
```

# The hue parameter

```python
sns.relplot(x='total_bill', y='tip', data=data, kind='scatter', hue='smoker')
```

# The style parameter

```python
sns.relplot(x='total_bill', y='tip', data=data, kind='scatter', hue='smoker', style='sex')
```

# The size parameter

```python
sns.relplot(x='total_bill', y='tip', data=data, kind='scatter', hue='smoker', style='sex', size='size', sizes=(10,200))
```

# Using the scatterplot function

```python
sns.scatterplot(x='total_bill', y='tip', data=data, hue='smoker', style='sex', size='size', sizes=(10,200))
```

# ***Video 4***

Violin plots<br>
Violin plot is a method of plotting numeric data and can be considered as a combination of box plot and kernel density plot (kde).

The unquestionable advantage of violin plot over the box plot is that aside from showing the above mentioned statistics, it also shows the entire distribution of the data.

Violin plots are ofter used to compare the distribution of a given variable across some categories.


```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=True, data=tips)
```

# **Video 8**

## jointplot
Joint plot creates a multi-panel figure that shows both the bivariate (joint) relationship between two variables along with the univariate (marginal) distribution of each on seperate axes.

We can use jointplot in the following 4 manners:
1. ScatterPlots
2. HexbinPlots
3. kdePlots
4. RegPlot

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

# Video 10
Distplot
Distplots are used to do univariate analysis on a continuous/numerical data.

Distplot combines 3 things in one graph.
1. Histogram
2. Kernel Density Estimation (PDF)
3. Rugplot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.describe()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

# Plotting a distplot

```python
sns.distplot(titanic['age'])
```

# Plotting only the histogram

```python
sns.distplot(titanic['age'], bins=10, kde=False)
```

# Plotting only the kde

```python
sns.distplot(titanic['age'], hist=False )
```

# Plotting only the rugplot

```python
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```

# How to use the distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'])
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])

```
