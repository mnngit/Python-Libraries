# Pie Chart

```python
sectors = ['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
```

```python
plt.pie(budget, labels=sectors, autopct="%0.2f%%")
plt.show()
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[200, 400, 700, 100, 200, 400, 140, 250]
chor_bazar=[100, 300, 400, 50, 100, 200, 70, 150]
```

```python
plt.plot(product_id, online_price, color='red', marker='D', markersize=10, label='Online Prices') # for 'marker' we can use D, +, etc
plt.plot(product_id, offline_price, color='green', markersize=10, label='Offline Prices')
plt.plot(product_id, chor_bazar, color='black', marker='+', markersize=10, label='Chor Bazar Prices')
plt.xlabel('Product')
plt.ylabel('Price')
plt.title("Product Price Comparison")
plt.legend()
plt.show()
```
